```python
import csv
import cv2
import numpy as np

def process_img(png, txt, folder):
    with open(txt) as f:
        your_list = []
        for line in f:
            line = line.split() # to deal with blank 
            if line:            # lines (ie skip them)
                line = [i for i in line]
                your_list.append(line)
        a=len(your_list)
        #for i in range (a):
            #del your_list[i][0]
        print(a)

        img = cv2.imread(png, cv2.IMREAD_COLOR).astype(np.uint8)
        
        for k in range(a):
            x = int(float(your_list[k][1])*size)
            y = int(float(your_list[k][2])*size)
            c= int(your_list[k][0])
            conf= int(float(your_list[k][5])*100)
            if x-75 < 0 or x+75 > size or y-75 < 0 or y+75 > size:
                continue
            crop_img = img[y-75:y+75, x -75: x +75].copy()
            if c==0:
                C="Inclusion"
            if c==1:
                C="Interface"
            if c==2:
                C="Martensite"
            if c==3:
                C="Notch"
            

            
            #path = folder + '/' + "class=" + str(C) + ", " + "Xcenter=" + str(x)+ ", " + "Ycenter=" + str(y) + ", "  + "confidence="+ str(conf) +"%"+ '.png'
            path = folder + '/' + str(C) + ", " + "Xcenter=" + str(x)+ ", " + "Ycenter=" + str(y) + ", "  + "confidence="+ str(conf) +"%"+ '.png'
            check = cv2.imwrite(path, crop_img)
            if check == True:
                print(path, " is ok.")
            else:
                print(path, " failed!")
```
